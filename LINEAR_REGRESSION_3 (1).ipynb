```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
#Load the dataset
df=pd.read_csv('delivery_time.csv')
```

```python
#Split the data into training and testing sets
X_train, X_test, y_train, y_test=train_test_split(df['Sorting Time'].values.reshape(-1,1), df['Delivery Time'].values, test_size=0.2, random_state=0)
```

```python
#Build the linear regression model
model=LinearRegression()
model.fit(X_train, y_train)
```

```python
#Make prediction
y_pred=model.predict(X_test)
```

```python
#Calculate RMSE
rmse=np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
#Visualize the regression line for original data
plt.scatter(X_test, y_test, label='Test Data')
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Regression Line')
plt.title(f'Linear Regression with original Data(RMSE:{rmse:2f})')
plt.xlabel('Original Sorting Time')
plt.ylabel('Delivery Time')
plt.legend()
plt.show()
```

```python
#Apply log transformation to the 'Sorting Time' feature
X_transformed_log=np.log1p(df['Sorting Time'].values.reshape(-1,1))
```

```python
#Split the data into training and testing sets for log transformation
X_train_log, X_test_log, y_train_log, y_test_log= train_test_split(X_transformed_log,df['Delivery Time'].values,test_size=0.2, random_state=0)
```

```python
#Build the linear regression model for log transformation
model_log=LinearRegression()
model_log.fit(X_train_log, y_train_log)
```

```python
#make prediction for log-transformed data
y_pred_log=model_log.predict(X_test_log)
```

```python
#Calculate RMSE for log-transformed data
rmse_log=np.sqrt(mean_squared_error(y_test_log,y_pred_log))
```

```python
#Visualize the regression line for log-tranformed data
plt.scatter(X_test_log, y_test_log, label='Test Data(Log Transformed)')
plt.plot(X_test_log, y_pred_log, color='red', linewidth=2, label='Regression Line(Log Transformed)')
plt.title(f'Linear Regression with log Transformation (RMSE:{rmse_log:.2f})')
plt.xlabel('Log Transformed sorting Time')
plt.ylabel('Delivery Time')
plt.legend()
plt.show()
```

```python
# Split the data into training and testing sets for square transformation
X_train_square, X_test_square, y_train_square, y_test_square = train_test_split(X_transformed_square, df['Delivery Time'].values, test_size=0.2, random_state=0)
```
